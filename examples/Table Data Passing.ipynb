```python
import weave
# weave.use_frontend_devmode()
```

```python
panel = weave.panels.Group(
    items={
        "table": weave.panels.Table(weave.ops.range(1, 100, 1), 
            columns=[
                lambda row: row,
                lambda row: row ** 2,
            ],
        ),
        "all_rows": lambda table:  weave.panels.Plot(table.all_input_rows(),
            x=lambda row: row['c_0'],
            y=lambda row: row['c_1']
        ),
        "derived": lambda table: weave.panels.Group(
            preferHorizontal=True,
            layoutMode="horizontal",
            items={
                "rows": weave.panels.Group(
                    items={
                        "pinned_rows": weave.panels.Plot(table.pinned_output_rows(),
                            x=lambda row: row['c_0'],
                            y=lambda row: row['c_1']
                        ),
                        "active_row": table.selected_output_row(),
                    }
                ),
                "data":  weave.panels.Group(
                    items={
                        "pinned_data": table.pinned_output_rows(),
                        "active_data": table.selected_output_row(),
                    }
                )
            })  
    })

panel
```
