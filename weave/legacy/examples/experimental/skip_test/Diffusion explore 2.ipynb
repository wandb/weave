```python
import weave
from weave.legacy.weave import weave_internal
from weave.legacy.weave.ecosystem.all import langchain
weave.use_frontend_devmode()
```

```python
weave.legacy.weave.panels.Board(
    vars={
        "chat_model": langchain.chat_openai("gpt-3.5-turbo", 0.7),
        "animals": weave.legacy.weave.panels.Query(
            ["koala", "giraffe", "leopard"],
            text=lambda item: item
        ),
        "objects": weave.legacy.weave.panels.Query(
            ["spoon", "guitar"],
            text=lambda item: item
        ),
        "prompts": weave.legacy.weave.panels.Query([
                "A {animal} with an {object}",
                "A 20-word story about a {animal} with an {object}",
                "You are creative writer. Describe a scene with a {animal} and a {object}. Choose an art style and use adjectives that evoke it."
            ],
            text=lambda item: item),
        "combos": lambda animals, objects, prompts:
            weave.legacy.weave.ops.cross_product(weave.legacy.weave.ops.dict_(
                animal=animals.pinned_data(),
                object=objects.pinned_data(),
                prompt=prompts.pinned_data())),

    },
    panels=[
        weave.legacy.weave.panels.BoardPanel(
            lambda chat_model, combos: weave.legacy.weave.panels.Facet(
                combos,
                cell_size=(256, 256),
                x=lambda item: item['animal'] + ' ' + item['object'],
                y=lambda item: item['prompt'],
                select=lambda items: 
                weave.legacy.weave.panels.Group(
                    vars={
                        'gen_text': chat_model.predict(
                            items[0]['prompt'].format(weave.legacy.weave.ops.dict_(
                                animal=items[0]['animal'],
                                object=items[0]['object'],
                            ))),
                    },
                    items={
                        'text': lambda gen_text: gen_text,
                        'image': lambda gen_text: gen_text.stable_diffusion()
                    }
                )
            ),
            layout=weave.legacy.weave.panels.BoardPanelLayout(x=0, y=0, w=24, h=24)
        )
#         weave.legacy.weave.panels.BoardPanel(
#             lambda col_names: weave.legacy.weave.panels.Table(col_names, columns=[lambda col: col]),
#             layout=weave.legacy.weave.panels.BoardPanelLayout(x=0, y=12, w=12, h=6)
#         ),
#         weave.legacy.weave.panels.BoardPanel(
#             lambda row_names: weave.legacy.weave.panels.Table(row_names, columns=[lambda row: row]),
#             layout=weave.legacy.weave.panels.BoardPanelLayout(x=12, y=12, w=12, h=6)
#         ),
    ]
)
```
